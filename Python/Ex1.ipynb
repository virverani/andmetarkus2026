```python
## Funktsiooni harjutus

def funktsioon (aasta : int) -> str: 
    if aasta < 0: 
        return f"{abs(aasta)} B.C."
    else: 
        return f"{aasta} A.D."

funktsioon(-1000)
```

```python
## Ütle inimesele tere

inimesed = {'Kati', 'Mati', 'Silver', 'Hannes'}

for nimi in inimesed: 
    print(f"Tere, {nimi}!")
```

```python
## Anna inimese nimi koos järjekorranumbriga 

inimesed = {'Kati', 'Mati', 'Silver', 'Hannes'}

jrk = 0

for nimi in inimesed: 
    jrk = jrk + 1
    print(f"{jrk}. inimese nimi on {nimi}.")
```

```python
## Anna inimese nimi koos järjekorranumbriga - alternatiivne lahendus.

inimesed = {'Kati', 'Mati', 'Silver', 'Hannes'}

jrk = 0

for nimi in inimesed: 
    jrk += 1
    print(f"{jrk}. inimese nimi on {nimi}.")
```

```python
## Kontrolli, kui palju tähti on nimes

inimesed = {'Kati', 'Mati', 'Silver', 'Hannes'}

jrk = 0

for nimi in inimesed: 
    jrk += 1
    if len(nimi) > 4:
        print(f"Pikk nimi: {nimi}.")
    else: 
        print(f"Lühike nimi: {nimi}.")
```

```python
## Lisa kaks sõna lausesse

inimesed = ['Kati', 'Mati', 'Silver', 'Hannes']
numbrid = ['Esimene', 'Teine', 'Kolmas', 'Neljas']

jrk = 0

for nimi in inimesed: 
    if len(nimi) > 4:
        print(f"{numbrid[jrk]} inimene kannab pika nime: {nimi}.")
    else: 
        print(f"{numbrid[jrk]} inimene kannab lühikest nime: {nimi}.")
    jrk += 1
```
